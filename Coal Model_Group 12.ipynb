In order to build this model, we focused on temporal prediction in order to enable prediction of coal job loss or solar job gains in each state by year. In order to do this, we had to use a timelag of 1 year - if our prediction year was 2013, our features were pulled from 2012 data, 2014 with 2013, and so on. Important to note here is that some of our features included forecasted capacity additions and retirements by fuel source and that data was also pulled from the same year as the features. In other words, forecasts made in 2012 about capacity changes in subsequent years (we set it to be 7 years out) were also factored into our 1 year future predictions for 2013. This was done for each subsequent year and while we were able to use historical data for most of the training and testing for our model from years 2013-2021, we also had to use forecasted data for some features in our 2022 prediction given that we didnt have historical values for features like income, population, and educational statistics by state. To do this forecasting, we used the forecasting tool in Excel which also ran a linear regression to predict forecasted values based on previous trends. We understand that this makes our results potentially less credible but we believe that it can still provide a rough estimate of the features that we need in order to predict our y-variable (jobs). 

**1.4 INPUT DATA DESCRIPTION** 

The data that we used to build our coal model included the following - 
- Bureau of Labor Statistics Data for historic employment (2012-2020) in the coal industry
- Energy Information Data from 2012-2020 on nameplate capacity from the previous year, as well as projected capacity, additions and retirements by state from the previous year. In order to find this data we used the Annual Energy Outlook, the Short Term Energy Outlook, Electric Power Monthly, Electric Power Annual, reported out from the EIA-860 forms. 
- Lazard data on levelized costs of energy for coal from 2012-2021. We assumed a constant LCOE for each state based on the yearly value. 
- American Community Survey and Historic US Census Data for demographic features including median household income, resident population, population above 25 years, population with educational attainments reaching high school, population with educational attainments reaching an associates degree, and population with educational attainments reaching a bachelor's degree. 

**SGSTF of Data Listed Above:** 
- **Structure:** For each of our individual data sets (which we merged into one dataframe) the data
    is grouped by state and fuel source. Each observation in the data frame corresponds to a year and a state, and each column represents either our response variable data or our feature variable data for that state in that year. We downloaded all of these files as Excel files.
    
- **Granularity:** Each row represents an observation from a state in each year from 2012-2020. 

- **Scope:** For our geographic scope, our data covers all states of the United States except for Washington DC, Rhode Island and Vermont which we ultimately removed from our model building process because they had no data relevant to the model's ability to predict - there were no coal jobs in these states, nor any coal capacity, nor any projections for additions or retirements. 

- **Temporality:** Our data set covers the years between 2012-2020, and we predicted for the years 2021 and 2022. All predictions were done by state.

- **Faithfulness:** Our demographic data comes from the Census Bureau, which is a government group that collects self-reported data on the population. Given that the data is self reported and voluntarily recorded, our data may be potentially skewed. Our BLS data comes from an annual mail survey of establishments that provides data on employment and wages by occupation and industry for over 800 occupations and for about 400 industries throughout the United States, and similar data for all states and selected metropolitan areas. Similar to the census data, there is potential for the data to be skewed because it is self reported and voluntarily recorded. However, the coal industry jobs data has a history of reporting to the BLS and seems reliable. In contrast, we couldn't use the solar and wind job data from BLS and had to turn to other sources (and exclude a wind model) because the jobs data was underreported for those sectors. Our Lazard data comes from historical averages and cross-validated with other LCOE estimates - little opportunities for data skewing. Finally, our EIA data came from multiple different datasets within the EIA database. Historic nameplate capacity data (for the previous year from our projection year) came from Electric Power Annual and our projected capacity additions and retirements came from the December version of Electric Power Monthly for years 2012-2020. For 2021 data (projecting 2022), we used the latest version of Electric Power Monthly which was released in October. We understand that this is an inconsistency in our data but unfortunately due to the fact that the December version has yet to be released yet, we had to rely on this data for our predictions. All of this data was originally downloaded catagorized by power plant (Form EIA-860) but we did some work in Excel first to sum the capacity, additions and retirements by state througn the use of a pivot table. This data is of course subject to change but because it is based on mandatory reporting at the power plant level, the data for the most part seems to be quite accurate in assessing coal capacity in the country.

Because there were multiple datasets that we needed to collate for each variable and because each dataset had a different granularity and temporality of features, for example EIA data being catagorized by power plant and demographic data by state, we had to prep a lot of our datasets first in excel in order to get all variables defined by state and year.
```python
from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()
loo.get_n_splits(X)
y_tests = []
y_predictions = []

for train, test in loo.split(X): #a
    Xr_train, Xr_test = np.array(X)[train], np.array(X)[test] #b
    yr_train, yr_test = np.array(y)[train], np.array(y)[test] #c
    
    lr = LinearRegression()
    lr.fit(Xr_train, yr_train) 
    yr_pred = lr.predict(Xr_test) #d
    
    y_tests.append(yr_test[0][0]) #e
    y_predictions.append(yr_pred[0][0]) #e

r2_loo = r2_score(y_tests, y_predictions)
MSE_loo = np.exp(mean_squared_error(yr_test, yr_pred))
print("R^2 after LOOCV: ", r2_loo)
print("OLS MSE after LOOCV: ", np.exp(MSE_loo))
```

```python
kf = KFold(n_splits = 4, shuffle = True, random_state = 1)

fold_mse = [] # initiate a list to hold the MSE associated with each fold

for t_index, v_index in kf.split(X):
    # Subset X and y into training and validation subsets
    X_fold_train = np.array(X)[t_index]
    y_fold_train = np.array(y)[t_index]
    X_fold_val = np.array(X)[v_index]
    y_fold_val = np.array(y)[v_index]
    # Initiate and fit a linear regression model using the training data
    lm = LinearRegression()
    lm.fit(X_fold_train, y_fold_train)
    # Predict the Y-values associated with the validation data
    y_pred = lm.predict(X_fold_val)
    
    # Find the testing MSE and append it to fold_mse
    MSE_F = mean_squared_error(y_fold_val, y_pred)
    fold_mse.append(MSE_F) 
    
print(fold_mse)
```

```python
def fit_model(SKL_model, X_train, X_test, y_train, y_test, alpha = 1):
    
        
        
    if SKL_model==LinearRegression():
        model != SKL_model()
        model.fit(X_train, y_train)
        model.coefficients = model.coef_
        y_pred = model.predict(X_test)
    else:
        model = SKL_model(alpha)
        model.fit(X_train, y_train)
        model.coefficients = model.coef_
        y_pred = model.predict(X_test)
           
    
    mse = mean_squared_error(y_test, y_pred)
    coef = model.coef_.flatten() 
    
    return mse, coef
```

## Interpretation and Conclusions

Our results indicate that our feature variables have some predictive power to predict our response variables. It is important to first address the fact that set our y variable array was defined as the log of our response variables because many of our data inputs had 0s thus affecting our distribution (seen in the histogram above). Our data had many zeros in part because there weren't projections of capacity additions and retirements for every state and there also aren't coal jobs in every state thus affecting the distribution of our response variable. When getting our results for our predictions for job losses in the coal industry in 2021 and 2022, we had to first take the exponent of our values to cancel out the log but in doing so, some of our predicted values for jobs are exactly or around the value of 1 in states where there weren't any coal jobs as far back as 2014. This is because the log(0) is 1 and so when interpreting the results it is important to caveat our predictions by noting that job numbers equaling around 1 can be counted as zero. That being said, it is clear that there are a handful of states with a significant number of employees in the coal industry and that those jobs, according to our model, are declining quite rapidly in places like Pennsylvania and Kentucky. It was also interesting to see that some states didn't have coal jobs but had coal nameplate capacity - we believe this is due to the fact that some states have coal in their electricity generation portfolio but aren't necessarily producing coal in their state but rather importing it from other parts of the country. 

Based on our model tests, as outlined above, we chose to conduct our predictions using a cross-validated Ridge model because the test score was highest when comparing all of the parametric models that we tested. Using the Ridge model, we are able to see that job losses in the coal industry are going to continue to happen at a rapid pace. As such, it is important now more than ever to allocate resources towards a just transition, with resources in particular to support transitioning workers out of the industry. Without that intentional allocation of resources, rapid job loss in the coal industry will only continue, creating economic dislocation for concentrated coal communities. We would recommend that policymakers at a state and federal level invest resources in these communities and also provide them with opportunities in either new energy industries (discussed further in the solar model) or broader opportunities for regional economic diversification. Currently, very few resources are allocated towards just transition policies in the United States but with our model, understanding the timeframe and rapidity of potential coal job loss, policymakers could begin to conduct proactive planning around supporting coal-dependent communities. We hope that readers will also care about our results, understanding that unfortunately the economics of the coal industry are ultimately not competitive with newer technologies. 

Our results should be taken into account with a few caveats: first, our data was based on historical as well as forecasts for the future. While the EIA is a trusted source for these types of projections, some of the additions and retirements that are projected by the Agency are not necessarily guaranteed to be retired (or added). Second as discussed in the data sections above, our demographic data and BLS data is based on self-surveys which automatically means potential skewing of the data. Moreover, for our 2022 predictions we also had to use Excel forecasted values of demographics for 2021. However, all that being said our data does clearly have some significant relationship because all of our model testing had high validation scores across different types of models - OLS, Ridge and Lasso, and all of the tree models. Moreover, the model is not particularly externally valid because it is unclear whether the model could be applied to another scope or dataset. Our model also falls within the bias variance tradeoff by potentially having high variance. However, it is interesting to note that both our solar and coal models utilized a cross-validated Ridge model, potentially suggesting that our model does have some external validity when applied to other datasets using the same types of features.
